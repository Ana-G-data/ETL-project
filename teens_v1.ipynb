```python
cursos=df["Course"].unique()
cursos
```

```python
indice = range(len(cursos))
serie_cursos = pd.Series(cursos).reset_index(drop=True)
serie_cursos
```

```python
def dividir_serie(serie):
    # Crear un DataFrame vac√≠o
    df_cursos = pd.DataFrame(columns=['Destino', 'Tipo', 'Duraci√≥n', 'Comienzo', 'Final', 'Nombre_completo'])
    
    # Recorrer cada elemento de la serie
    for i, elemento in serie_cursos.iteritems():
        # Extraer las partes de la cadena usando expresiones regulares
        match = re.search(r'(.*) (TEENS|ADULTS)-VOLUNTEER (.*) from (.*) to (.*)', elemento)
        destino = match.group(1)
        tipo = match.group(2)
        duracion = match.group(3)
        comienzo = match.group(4)
        final = match.group(5)
        
        # A√±adir las partes extra√≠das y el nombre completo al DataFrame
        df_cursos.loc[i] = [destino, tipo, duracion, comienzo, final, elemento]
    
    return df_cursos

# Aplicar la funci√≥n a la serie
df_cursos = dividir_serie(serie_cursos)
df_cursos
```

```python
df_arrival_plane=df[df['Arrival_Medium']== 'Plane']
df_arrival_plane.shape
```

```python
ciudad_origen=df_arrival_plane["Origin_City"].unique()
ciudad_origen
```

```python
def extraer_informacion(ciudad_original):
    # Primero buscamos el nombre de la ciudad despu√©s de la primer coma
    nombre_ciudad = ciudad_original.split("-")[0]
    # Si no hay coma, el nombre de la ciudad es la cadena original
    if not nombre_ciudad:
        nombre_ciudad = ciudad_original
    
    # Luego buscamos el c√≥digo IATA al principio o al final de la cadena, entre par√©ntesis o despu√©s de un gui√≥n
    codigo_iata = re.search(r"^([A-Z]{3})\)?|([A-Z]{3})\)?$", ciudad_original)
    if codigo_iata:
        codigo_iata = codigo_iata.group(1) or codigo_iata.group(2)
    
    return nombre_ciudad, codigo_iata

ciudades = []
for ciudad_original in ciudad_origen:
    nombre_ciudad, codigo_iata = extraer_informacion(ciudad_original)
    ciudades.append({
        "ciudad": nombre_ciudad,
        "codigo_iata": codigo_iata,
        "ciudad_original": ciudad_original,
    })

df_ciudades = pd.DataFrame(ciudades)
```

```python
df_iata = pd.read_csv('datos/codigo_iata.csv')
df_iata.head()
```

```python
from fuzzywuzzy import fuzz

# Iterar sobre las filas del dataframe df_ciudades
for index, row in df_ciudades.iterrows():
  ciudad = row['ciudad']
  max_similarity = 0
  match = None
  
  # Iterar sobre las filas del dataframe df_iata
  for index_iata, row_iata in df_iata.iterrows():
    sirve_a = row_iata['Sirve a']
    
    # Calcular el parecido entre las dos cadenas de texto
    similarity = fuzz.token_set_ratio(ciudad, sirve_a)
    
    # Si el parecido es mayor que 50 y mayor que el m√°ximo encontrado hasta ahora, actualizar la variable temporal
    if similarity > 80 and similarity > max_similarity:
      max_similarity = similarity
      match = sirve_a
  
  # Si se ha encontrado una coincidencia con parecido mayor que 50, actualizar la columna ciudad_origen en el dataframe df_ciudades
  if match is not None:
    df_ciudades.loc[index, 'ciudad'] = match
```
