```python
import requests
from datetime import datetime, time, timedelta

API_KEY = token

# Set the base URL for the API
BASE_URL = "http://api.aviationstack.com/v1/flights"

# Set the parameters for the API call
params = {
    "access_key": API_KEY,
    "airline_iata": "TK",  # Iberia's airline code
    "dep_iata": "DFW",  # Dallas airport code
    "arr_iata": "MAD",  # Madrid airport code
    "arr_date": datetime.now().strftime("%Y-%m-%d"),  # Arrival date should be today
}

# Make the API call
response = requests.get(BASE_URL, params=params)

# Check the status code of the response
if response.status_code == 200:
    # If the request is successful, the flight information will be in the "data" field of the response
    data = response.json()["data"]
    print(data)
else:
    # If the request is not successful, print the status code and error message
    print(f"Request failed with status code {response.status_code}: {response.text}")
```

```python
def get_flight_arrival_details(flights):
  # Set the API key and base URL for the Aviation Stack API
  api_key = token
  base_url = "http://api.aviationstack.com/v1/flights"

  # Initialize an empty list to store the flight details
  flight_details = []

  # Iterate over the flights dictionary
  for flight_number, arrival_date in flights.items():
    # Set the parameters for the API request
    params = {
      "flight_iata": flight_number,
      "arr_date": arrival_date,
      "arr_iata": "MAD",
      "access_key": api_key
    }

    # Make the request to the Aviation Stack API
    response = requests.get(base_url, params=params)

    # Check the status code of the response
    if response.status_code == 200:
      # If the request is successful, extract the flight details from the response
      
      flight_details.append(response.json())
    else:
      # If the request is unsuccessful, print an error message
      print(f"Failed to get details for flight {flight_number}: {response.status_code}")

  return flight_details
```

```python
flights = {
  "IB3405": "2022-12-27",
  "FR6758": "2022-12-27",
  "TK1859": "2022-12-27"
}

flight_details = get_flight_arrival_details(flights)
print(flight_details)
```

```python
def sacar_info(col):
    '''
    Extrae la información que necesitamos de un diccionario
    Args: 
        col (str): el nombre de la columna sobre la que queremos aplicar la función
    Returns: 
        el diccionario con la información que queremos
    '''
    try:
        return  col[0]["arrival"]
    except: 
        pass
```

```python
df_flight_details["Details"] = df_flight_details["data"].apply(sacar_info)
```
